#### Write a Python application to scrape the data from the following web site as per the directions given.
http://quotes.toscrape.com/

Data from the first 5 pages of this site need to be scraped.

### Importing libs

```python
#to use the sleep feature
import time

#DataFrames, Series Data Mauplation
import pandas as pd

#Use the web browser like a human
from selenium import webdriver
```

>A)	The output of your Python code should be displayed as shown below.
> ---

```python
#We are going to use Firefox as our Browser
browser = webdriver.Firefox()

#For iterating through pages
pageno = 1 

#Original Site url
url = "http://quotes.toscrape.com/"

Quotes = []

Authors = []

Tags = []

while pageno <=5:
    # Selecting page after page untill 5th page
    pageurl = str(url+'page/{}/'.format(pageno))
    
    #Now lets access FireFox and open our wesite using .get()
    browser.get(pageurl)
    
    #Scraping quotes
    for quoteno in range(1, 11):
        #Quote
        quotepath = ('/html/body/div/div[2]/div[1]/div[{}]/span[1]'.format(quoteno))
        
        Q = browser.find_element_by_xpath(quotepath)
        
        Quotes.append(Q.text[1:-1])
        
        print('Quote:', Q.text[1:-1])
        
        #Author
        authorpath = ('/html/body/div/div[2]/div[1]/div[{}]/span[2]/small'.format(quoteno))
        
        A = browser.find_element_by_xpath(authorpath)
        
        Authors.append(A.text)
        
        print('Author:', A.text)
        
        #Tags
        tagspath = ('/html/body/div/div[2]/div[1]/div[{}]/div'.format(quoteno))
        
        T = browser.find_element_by_xpath(tagspath)
        
        Tags.append(T.text.replace(' ', ';')[6:])
        
        print('Tags:', T.text.replace(' ', ';')[6:])
    
    #time.sleep(4)
    pageno += 1;
    
```

>B)	The scraped data is to be loaded into a Pandas dataframe and written into a CSV file.
> ----

```python
Quote_df = pd.DataFrame({'Author' : Authors, 'Quote' : Quotes, 'Tags' : Tags})

Quote_df.head()
```

```python
Quote_df.to_csv('Quotes_scraped.csv')
```

#### C)	As part of each quote a hyperlink is also present as highlighted in the screen shot below. It points to a web page about the author.

> #### Modify your code to extract the hyperlink as well and display it as shown below.

```python
#We are going to use Firefox as our Browser
browser = webdriver.Firefox()

#For iterating through pages
pageno = 1 

#Original Site url
url = "http://quotes.toscrape.com/"

Quotes = []

Authors = []

Tags = []

About = []

while pageno <=5:
    
    # Selecting page after page untill 5th page
    pageurl = str(url+'page/{}/'.format(pageno))
    
    #Now lets access FireFox and open our wesite using .get()
    browser.get(pageurl)
    
    #Scraping quotes
    for quoteno in range(1, 11):
        #Quote
        quotepath = ('/html/body/div/div[2]/div[1]/div[{}]/span[1]'.format(quoteno))
        
        Q = browser.find_element_by_xpath(quotepath)
        
        Quotes.append(Q.text[1:-1])
        
        print('Quote:', Q.text[1:-1])
        
        #Author
        authorpath = ('/html/body/div/div[2]/div[1]/div[{}]/span[2]/small'.format(quoteno))
        
        A = browser.find_element_by_xpath(authorpath)
        
        Authors.append(A.text)
        
        print('Author:', A.text)
        
        #About
        
        aboutpath = ('/html/body/div/div[2]/div[1]/div[{}]/span[2]/a'.format(quoteno))
        
        Ab = browser.find_element_by_xpath(aboutpath)
        
        About.append(Ab.get_attribute('href'))
        
        print('About:', Ab.get_attribute('href'))
        
        #Tags
        tagspath = ('/html/body/div/div[2]/div[1]/div[{}]/div'.format(quoteno))
        
        T = browser.find_element_by_xpath(tagspath)
        
        Tags.append(T.text.replace(' ', ';')[6:])
        
        print('Tags:', T.text.replace(' ', ';')[6:])
        

    #time.sleep(4)
    pageno += 1;
```

> #### Similarly the Python program needs to save the link in the dataframe as a column (series) and save it to the CSV file also.

```python
Quote_df_with_about = pd.DataFrame({'About' : About, 'Author' : Authors, 'Quote' : Quotes, 'Tags' : Tags})  

Quote_df_with_about.head()
```

```python
Quote_df_with_about.to_csv('Quotes_with_about.csv')
```
