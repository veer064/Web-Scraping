```python
%%time
#We are going to use Firefox as our Browser
browser=webdriver.Firefox()

#Now lets access FireFox and open our wesite using .get()
browser.get('https://pgd.datascience.manipal.edu/?q=MULNLogin/dslogin')

#Giving User-id remotely
emailElem=browser.find_element_by_id('edit-loginId-1')
emailElem.send_keys('19225760060')

#Giving Password remotely
passElem=browser.find_element_by_id('edit-password-1')
passElem.send_keys('19225760060')

#Submit to enter into the website
emailElem.submit()
time.sleep(5) #sleep to make the browser feel like a real user

#entering into MyCourse
menu=browser.find_element_by_xpath('/html/body/div[2]/header/div/div/div/div/div[1]/div[2]/nav/div/ul/li[3]')
menu.click()

#selecting subjects
subjects={4:'pds', 5:'stds', 6:'dsdw', 7:'dav', 8:'bdt', 9:'ml'}

#List of DataFrame's names
df_lists=[]

#Final DataFrame with Names, Roll.no., and all subjectsject percentages of Assesments and course completion.
scores_df=pd.DataFrame()


for k in range(4,10):
    #Subject container for each 'k'
    xpath=('//*[@id="sub-container"]/a[{}]/div').format(k)
    sub=browser.find_element_by_xpath(xpath)
    sub.click()
    time.sleep(6)
    
    #Once enter into Subject page, we need to click on performance
    perf=browser.find_element_by_xpath('//*[@id="studentReports"]')
    perf.click()
    time.sleep(6)
    
    #Then our destination, Leader Board
    lederb=browser.find_element_by_xpath('//*[@id="content-content"]/div/div/div/div[2]/div/ul/li[3]')
    lederb.click()
    time.sleep(6)
    
    #Changing the no.of. rows to show option to 'All'
    sel=browser.find_element_by_xpath('/html/body/div[2]/section/div/div/div[2]/div/div/div/div/div/div/div[2]/div/div/div/div[2]/div/div/div[1]/label')
    sel.click()
    op=browser.find_element_by_xpath('//*[@id="leaderboard-students-list_length"]/label/select/option[5]')
    op.click()
    
    #selecting columns: 2, 3, 4, and 5
    for j in range(2,6):
        globals()['list_{}_{}'.format(k,j)]=[]
        for i in range(1,18):
            xpath=('//*[@id="leaderboard-students-list"]/tbody/tr[{}]/td[{}]').format(i,j)
            
            #extracting data from each column
            data=browser.find_element_by_xpath(xpath)
            
            #Data from each columns falls under new list respectively
            globals()['list_{}_{}'.format(k,j)].append(data.text)
        
    #Back to MyCourse menu, as our work on the current subject leader board is done.        
    menu=browser.find_element_by_xpath('//*[@id="secondary-menu"]/ul/li[3]')
    menu.click()

 
    
    #List for each subject, containing the 4 listed created in the above 
    globals()["{}".format(subjects[k])]=[]
    
    
    ##Appending those 4 list into the subject wise list created right above 
    for j in range(2,6):
        globals()["{}".format(subjects[k])].append(globals()['list_{}_{}'.format(k,j)])
    
    #Creating Sujectwise DataFrame from the above create subject wise list
    globals()["{}_df".format(subjects[k])]=pd.DataFrame(globals()["{}".format(subjects[k])]).transpose()
    globals()["{}_df".format(subjects[k])].columns=['Name','Roll_Number','{}_Ass_Pct'.format(subjects[k]),'{}_Course_Pct'.format(subjects[k])]
    
    #Saving the individual subject DataFrame to csv on our machine
    globals()["{}_df".format(subjects[k])].to_csv('{}.csv'.format(subjects[k]))
    
    #Appending all subject wise DataFrame Names created in the above
    #For our refference
    df_lists.append("{}_df".format(subjects[k]))
    df_lists
```
